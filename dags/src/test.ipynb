```python
import os
import json
from typing import Any

def determine_seniority(months_of_experience):
    if months_of_experience < 25:
        return "Junior"
    elif months_of_experience < 61:
        return "Mid-Level"
    elif months_of_experience < 121:
        return "Senior"
    else:
        return "Lead/Principal"

def process_salary(json_file):
    salary_info = json_file.get("estimatedSalary", {}).get("value", {})
    return {
        "currency": json_file.get("estimatedSalary", {}).get("currency", ""),
        "min_value": salary_info.get("minValue", ""),
        "max_value": salary_info.get("maxValue", ""),
        "unit": salary_info.get("unitText", "")
    }

def transform_data(json_file):
    return {
        "job": {
            "title": json_file.get("title", ""),
            "industry": json_file.get("industry", ""),
            "description": json_file.get("description", ""),
            "employment_type": json_file.get("employmentType", ""),
            "date_posted": json_file.get("datePosted", ""),
        },
        "company": {
            "name": json_file.get("hiringOrganization", {}).get("name", ""),
            "link": json_file.get("hiringOrganization", {}).get("sameAs", ""),
        },
        "education": {
            "required_credential": json_file.get("educationRequirements", {}).get("credentialCategory", ""),
        },
        "experience": {
            "months_of_experience": json_file.get("experienceRequirements", {}).get("monthsOfExperience", ""),
            "seniority_level": determine_seniority(transform_data(json_file).get("experience", {}).get("months_of_experience", "")),
        },
        "salary": process_salary(json_file),
        "location": {
            "country": json_file.get("jobLocation", {}).get("address", {}).get("addressCountry", ""),
            "locality": json_file.get("jobLocation", {}).get("address", {}).get("addressLocality", ""),
            "region": json_file.get("jobLocation", {}).get("address", {}).get("addressRegion", ""),
            "postal_code": json_file.get("jobLocation", {}).get("address", {}).get("postalCode", ""),
            "street_address": json_file.get("jobLocation", {}).get("address", {}).get("streetAddress", ""),
            "latitude": json_file.get("jobLocation", {}).get("latitude", ""),
            "longitude": json_file.get("jobLocation", {}).get("longitude", ""),
        },
    }


def read_json(
    file_path: str,               # The path of the text file containing JSON
    encoding: str = "utf-8"       # The encoding of the file (default is UTF-8)
) -> Any:
    """
    Reads a text file containing JSON and returns its content.

    Args:
        file_path (str): The path to the text file containing JSON.
        encoding (str): The encoding of the file.

    Returns:
        Any: The content of the JSON file.
    """
    with open(file=file_path, mode="r", encoding=encoding) as f:
        # Use json.load to load the content of the JSON file
        return json.load(f)

def create_file_dir(
    file_dir: str
) -> None:
    """
    Creates a file dir if doesn't exist.

    Args:
        file_dir: the directory of the file
    
    Returns:
        None
    """
    if not os.path.isdir(file_dir):
        os.makedirs(file_dir)


for i, file_name in enumerate(os.listdir("../../staging/extracted/context_data")):
    if i<6:
        file_path = f"../../staging/extracted/context_data/{file_name}"
        print(file_name)
        data = read_json(file_path)
        # Convertir en chaîne JSON
        json_result = json.dumps(transform_data(data), indent=2)

        # Afficher le résultat
        print(json_result)
        # write_json(f"{OUTPUT_DIR_TRANSFORMED}/context{i}.txt", transform_data(data))
        i+=1
```
